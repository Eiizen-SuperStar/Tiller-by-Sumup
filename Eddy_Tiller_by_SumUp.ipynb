```python
# Importez les bibliothèques nécessaires
from google.cloud import bigquery
from google.colab import auth

# Authentifiez-vous avec votre compte Google
auth.authenticate_user()

# Initialisez le client BigQuery
project_id = "enduring-button-402908"
client = bigquery.Client(project=project_id)

# Spécifiez la requête SQL pour extraire les données
query = """
SELECT *
FROM `enduring-button-402908.tiller.order_data`
"""

# Exécutez la requête et récupérez les résultats dans un DataFrame Pandas
df_order_data = client.query(query).to_dataframe()

# Affichez les premières lignes du DataFrame
df_order_data
```

```python
# Spécifiez la requête SQL pour extraire les données
query = """
SELECT *
FROM `enduring-button-402908.tiller.order_line`
"""

# Exécutez la requête et récupérez les résultats dans un DataFrame Pandas
df_order_line = client.query(query).to_dataframe()

# Affichez les premières lignes du DataFrame
df_order_line
```

```python
# Spécifiez la requête SQL pour extraire les données
query = """
SELECT *
FROM `enduring-button-402908.tiller.payment_data`
"""

# Exécutez la requête et récupérez les résultats dans un DataFrame Pandas
df_payment_data = client.query(query).to_dataframe()

# Affichez les premières lignes du DataFrame
df_payment_data
```

```python
# Spécifiez la requête SQL pour extraire les données
query = """
SELECT *
FROM `enduring-button-402908.tiller.store_data`
"""

# Exécutez la requête et récupérez les résultats dans un DataFrame Pandas
df_store_data = client.query(query).to_dataframe()

# Affichez les premières lignes du DataFrame
df_store_data
```

```python
# Vérifiez les valeurs nulles dans df_order_data
print("Valeurs nulles dans df_order_data :")
print(df_order_data.isnull().sum())

# Vérifiez les valeurs nulles dans df_order_line
print("\nValeurs nulles dans df_order_line :")
print(df_order_line.isnull().sum())

# Vérifiez les valeurs nulles dans df_payment_data
print("\nValeurs nulles dans df_payment_data :")
print(df_payment_data.isnull().sum())

# Vérifiez les valeurs nulles dans df_store_data
print("\nValeurs nulles dans df_store_data :")
print(df_store_data.isnull().sum())
```

```python
df_order_line = df_order_line.rename(columns={'date_created': 'date_created_order_line'})

df_payment_data = df_order_line.rename(columns={'date_created': 'date_created_payment_data'})

df_store_data = df_order_line.rename(columns={'date_created': 'date_created_store_data'})

df_order_line.head()
```

```python
import pandas as pd

# Joignez df_order_data et df_order_line sur la colonne 'order_id'
merged_df = pd.merge(df_order_data, df_order_line, on='id_order', how='left')

# Affichez les premières lignes du DataFrame résultant
merged_df
```

```python
merged_df = pd.merge(merged_df, df_payment_data, on='id_order', how='left')

merged_df
```

```python
merged_df = pd.merge(merged_df, df_store_data, on='id_store', how='left')

merged_df
```

```python
# Liste des colonnes inutiles à supprimer
colonnes_inutiles = ["dim_category_translated", "dim_name_translated"]

# Supprimer les colonnes inutiles du DataFrame fusionné
merged_df = merged_df.drop(colonnes_inutiles, axis=1)
```
